```python
import requests
import pandas as pd
```

```python
def getdata(stock, apikey="616f4e72aaf4346ee78c7219ccbea7bf"):
 # Company Quote Group of Items
    company_quote = requests.get(f"https://financialmodelingprep.com/api/v3/quote/{stock}?apikey={apikey}")
    company_quote = company_quote.json()
    share_price = float("{0:.2f}".format(company_quote[0]['price']))
    
    # Income Statement Group of Items
    IS = requests.get(f"https://financialmodelingprep.com/api/v3/financials/income-statement/{stock}?apikey={apikey}&period=quarter")
    IS = IS.json()

    # Most Recent Quarterly Revenue
    qRev = float("{0:.2f}".format(float(IS['financials'][0]['Revenue'])/10**9))
    
    # Balance Sheet Group of Items    
    BS = requests.get(f"https://financialmodelingprep.com/api/v3/financials/balance-sheet-statement/{stock}?apikey={apikey}&period=quarter")
    BS = BS.json()

    #Total Debt
    debt = float("{0:.2f}".format(float(BS['financials'][0]['Total debt'])/10**9))
    #Total Cash
    cash = float("{0:.2f}".format(float(BS['financials'][0]['Cash and short-term investments'])/10**9))
    
    # Company Profile Group of Items
    company_info = requests.get(f"https://financialmodelingprep.com/api/v3/company/profile/{stock}?apikey={apikey}")
    company_info = company_info.json()
    # Chief Executive Officer
    ceo = company_info['profile']['ceo']
    
    
    return [share_price, qRev, debt, cash, ceo]
```

```python
price, rev, debt, cash, ceo = getdata('AAPL')

dataDict = {'price': price, 'revenue': rev, 'debt': debt, 'cash flow':cash, 'CEO':ceo}
```

```python
tickers = ['AAPL']

df = pd.DataFrame(dataDict, index=tickers)
```
